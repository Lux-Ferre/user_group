```python
# Step 1. Create list of abundants

list_abundants = []
for i in range(1,28123):
    # calculate the divisors (except itself)
    list_divisors=[]
    for j in range(1,i):
        if i%j==0 and i!=j:
            list_divisors.append(j)
    # sum the divisors & check if the sum is greater than the original
    if sum(list_divisors)>i:
        list_abundants.append(i)
        print(i)
print(list_abundants)
```

```python
# Step 2: Find the sum

list_final = []
for i in range(1,28123):
    # iterate the list of abundants
    flag = 0
    for j in list_abundants:
        if i>j:
            if i-j in list_abundants:
                flag = 1
    if flag==0:
        list_final.append(i)
        print(i)
print("The sum of all the positive integers which cannot be written as the sum of two abundant numbers is", sum(list_final))    
```
